```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data=pd.read_csv('./titanic.csv')
data.head()
```

### UNIVARIATE ANALYSIS

### KDE PLOT

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data.Pclass)
plt.show()
```

### HISTPLOT

```python
sns.histplot(data.Age)
plt.show()
```

### BOX PLOT

```python
sns.boxplot('Age',data=data)
plt.show()
```

### MULTIVARIATE ANALYSIS

### LINE PLOT

```python
sns.lineplot('Age','Fare',data=data)
plt.title("Age vs Fare")
plt.show()
```

```python
sns.lineplot('Age','Fare',data=data, hue="Sex")
plt.title("Age vs Fare")
plt.show()
```

```python
sns.lineplot('Age','Fare',data=data, hue="Pclass")
plt.title("Age vs Fare")
plt.show()
```

```python
sns.lineplot('Survived','Fare',data=data, hue="Pclass")
plt.title("Survived vs Fare")
plt.show()
```

```python
sns.lineplot('Survived','Fare',data=data, hue="Sex")
plt.title("Survived vs Fare")
plt.show()
```

### PIE CHART

```python
Pclass_Survived=data.groupby(['Pclass'])['Survived'].sum()
```

```python
sns.set_style('white')
Pclass_Survived.plot.pie()
plt.legend()
plt.show()
```

```python
Pclass_Sex_Survived=data.groupby(['Pclass','Sex'])['Survived'].sum()
```

```python
Pclass_Sex_Survived.reset_index()
```

```python
Pclass_Sex_Survived.plot.pie(autopct='%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,  1),loc='upper left',borderaxespad=0)
plt.show()
```

### BAR CHART

```python
plt.figure(figsize=(4,3))
sns.countplot('Sex',hue='Survived',data=data)
plt.show()
```

### COUNT PLOT

```python
plt.figure(figsize=(4,3))
sns.countplot('Pclass',data=data)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.set_style('darkgrid')
sns.countplot('Pclass', hue='Survived', data=data)
plt.title('Pclass: Survived vs Dead')
plt.show()
```

### FACTOR PLOT

```python
sns.set_style('ticks')
sns.factorplot('Pclass','Survived', hue='Sex', data=data)
plt.show()
```

### VIOLIN PLOT

```python
sns.set_style('dark')
plt.figure(figsize=(4,3.5))
sns.violinplot("Sex","Age", hue="Survived", data=data)
plt.show()
```

```python
sns.set_style('dark')
plt.figure(figsize=(4,3.5))
sns.violinplot("Pclass","Age", hue="Survived", data=data)
plt.show()
```

### BARPLOT

```python
plt.figure(figsize=(8,4))
sns.barplot('SibSp','Survived', data=data)
plt.title("SibSp & Survived")
plt.show()
```

### PAIR PLOT


```python
sns.pairplot(data=data)
plt.show()
```

### HEAT MAP

```python
heat_map=data.corr()
sns.heatmap(heat_map)
plt.show()
```
