#### PLOTING WITH PANDAS


```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("Car_sales.csv")
data.tail(10)
```

#### Droping na values

```python
data.dropna(inplace=True)
```

```python
data.info()
```

### Plots on numerical value columns

### Histogram

```python
data.Horsepower.plot.hist(figsize=(6,3), cmap='brg');
```

### Box plot

```python
data.Sales_in_thousands.plot.box(figsize=(4,3), cmap='gist_earth');
```

```python
data.Horsepower.plot.box(figsize=(4,3), cmap='Blues_r');
```

### Subplots

```python
data.plot.box(figsize=(22,4), subplots=True);
```

```python
data.plot.hist(figsize=(6,18),subplots=True);
```

### Line Chart

```python
data.sort_values(by='__year_resale_value').plot.line('__year_resale_value','Price_in_thousands',figsize=(6,3));
```

### Bar Plot

#### To create the below shown bar chart we have to create the frequency table. Frequency table can be made using the value_counts() method.

```python
data.Manufacturer.value_counts().plot.bar(figsize=(6,2), cmap='gnuplot2');
```

```python
data.sort_values(by='Price_in_thousands')[:20].plot.bar('Manufacturer','Price_in_thousands',figsize=(6,2),cmap='prism');
```

### Area Chart

```python
data.plot.area('Manufacturer','Price_in_thousands', figsize=(6,3), cmap='tab20b');
```

### Horizontal bar Chart

```python
data.sort_values(by='Price_in_thousands')[:5].plot.bar('Model','Power_perf_factor', figsize=(6,4));
```

### Scatter Plot

```python
data.sort_values(by='Price_in_thousands')[:5].plot.scatter('Latest_Launch','Price_in_thousands', figsize=(10,3));
```

```python
data.plot.box(figsize=(22,5));
```

```python
data.sort_values(by='Engine_size',ascending=False)[:10].Engine_size.plot.pie();
```

### KDE Plot

```python
data.plot.kde(figsize=(10,4));
```

```python
data.plot.kde(figsize=(12,10), subplots=True);
```

### Hexbin Plot

```python
data.plot.hexbin('Curb_weight','Fuel_capacity',gridsize=15, cmap='Pastel2');
```
